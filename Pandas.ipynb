```python
## Applying filters using loc
print (loan.loc[loan['Loan_id'] >= 30000000])                            ## where Loan_id >= 30000000
print (loan.loc[(loan['Loan_id'] >= 30000000) & (loan['Inter'] == 300)]) ## where Loan_id >= 30000000 and Inter = 300
print (loan.loc[(loan['Loan_id'] < 30000000) | (loan['Prin'] == 'Nan')]) ## where Loan_id < 30000000 and Prin is null
```

```python
## creating new dataframes from existing one based on multiple conditions;
## create table tb as select * from table_name where Loan_id >= 30000000 and Inter=100;
## Notice that the index positions aren't changed for new dataframe.
## Hence recommended to reset the indexes using reset_index

new_loan = loan.loc[(loan.Loan_id >= 30000000) & (loan.Inter == 100)]
print(new_loan.head())
#new_loan = new_loan.reset_index()
#new_loan = new_loan.reset_index(drop=True)
new_loan.reset_index(drop=True, inplace=True)
new_loan.head()
```

```python
## conditional filters viz., like; contains;
## adding a string column;
import re
loan['Static_Name'] = 'Static_Name'
print (loan.loc[loan['Static_Name'].str.contains('Name')]) ## select all columns where static_name like 'Name'
print (loan.loc[loan['Static_Name'].str.contains('static|name',regex=True,flags=re.I)]) ## selecting with regex
print (loan.loc[loan['Static_Name'].str.contains('St[a-z]*',flags=re.I,regex=True)]) ## again regex select
```

```python
## Conditional changes/updates
loan.loc[loan['Loan_id'] == 300000149, 'Loan_id'] = 100 ## update loan set Loan_id=100 where Loan_id = 300000149
print (loan.tail())
loan.loc[loan['Loan_id'] == 100, 300000149] = True
loan = loan.drop(columns=[300000149])
loan.tail()
```

```python
## multiple conditions and update
loan.loc[loan['Loan_id'] > 30000000, ['Inter', 'Prin']] = loan['Inter']+loan['Prin']
print (loan.tail())
loan.loc[loan['Loan_id'] > 30000000, ['Inter', 'Prin']] = [100, 200]
print (loan.tail())
```

```python
## Aggregating / Grouping
loan.groupby(['Loan_id']).mean() ## General group by loan_id and getting mean of all columns
loan.groupby(['Loan_id']).mean().sort_values('Inter', ascending=False) ## grouping and sorting based on mean of inter
```

```python
## sum() and count()
loan.groupby(['Loan_id']).sum() ## select sum(all columns) from table group by Loan_id
loan.groupby(['Loan_id']).count()
print (loan.head())
loan.groupby(['Loan_id','Inter']).count()['TotalAmt']
```
