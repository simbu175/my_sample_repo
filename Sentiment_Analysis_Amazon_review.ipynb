```python
## Checking the review dataset
product_review.head()
```

```python
product_review['Title_Polarity'] = ''
product_review['Title_Subjectivity'] = ''
product_review['Body_Polarity'] = ''
product_review['Body_Subjectivity'] = ''
product_review.head()
```

```python
product_review['Title_Polarity'] = product_review['ReviewTitle'].apply(lambda tweet: TextBlob(tweet).polarity)
product_review['Title_Subjectivity'] = product_review['ReviewTitle'].apply(lambda tweet: TextBlob(tweet).subjectivity)
product_review['Body_Polarity'] = product_review['ReviewBody'].apply(lambda tweet: TextBlob(tweet).polarity)
product_review['Body_Subjectivity'] = product_review['ReviewBody'].apply(lambda tweet: TextBlob(tweet).subjectivity)
product_review.head()
```
